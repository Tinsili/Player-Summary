```r
library(tidyverse)
```

```r
players <- read_csv("dsci-100-student/players.csv") |>
           select(experience, subscribe)
players
```

```r
players_summary <- players |>
                   group_by(experience, subscribe) |>
                   summarize(count = n())
players_summary
```

```r
players_summary_plot <- ggplot(players_summary, aes(x = experience, y = count, fill = subscribe)) +
                        geom_bar(stat = "identity")
players_summary_plot
```
